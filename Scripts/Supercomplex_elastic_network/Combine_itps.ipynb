```python
import glob

def general_shift(natoms):
    def fun(line, shift, resshift):
        *atoms, rest = line.split(maxsplit=natoms)
        atoms = ' '.join(str(int(i) + shift) for i in atoms)    
        return f'{atoms} {rest}'
    return fun


def do_exclusions(line, shift, resshift):
    atoms, *comment = line.split(';', maxsplit=1)
    if comment:
        comment = f' ; {comment[0].strip()}'
    else:
        comment = ''
    atoms = ' '.join(str(int(i) + shift) for i in atoms.split())
    return f'{atoms}{comment}\n'


def do_virtual_sites(line, shift,resshift):
    atoms, *comment = line.split(';', maxsplit=1)
    if comment:
        comment = f' ; {comment[0]}'
    else:
        comment = ''
    atoms = [int(i) + shift for i in atoms.split()]
    atoms[1] -= shift
    atoms = ' '.join(str(i) for i in atoms)
    return f'{atoms}{comment}\n'

def do_atoms(line, shift, resshift):
    atoms = line.split()
    for i in [0, 5]:
        atoms[i] = int(atoms[i]) + shift
    atoms[2] = int(atoms[2]) + resshift
    atoms = ' '.join(str(i) for i in atoms)
    return f'{atoms}\n'
##Need to shift the resdiue number seperatly 


do_shift = {
    'atoms': do_atoms, # general_shift(1),
    'bonds': general_shift(2),
    'constraints': general_shift(2),
    'angles': general_shift(3),
    'dihedrals': general_shift(4),
    'exclusions': do_exclusions,
    'virtual_sitesn': do_virtual_sites,
}


def renumber_itp(itpfile, shift=0, resshift=0):
    out = []
    atm = []
    bnd = []
    con = []
    dih = []
    ang = []
    exc = []
    vs = []
    tag = None
    natoms = 0
    with open(itpfile) as itp:
        for line in itp:
            stripped = line.strip()
            if stripped.startswith('['):
                tag = stripped[1:].split(']', maxsplit=1)[0].strip()
                if tag == 'moleculetype':
                    out.append(f'; [ {tag} ]\n')
                    out.append(f'; {next(itp)}\n')
                else:
                    out.append(line)
                continue
            if not tag or not stripped or stripped.startswith((';', '#')):
                out.append(line)
                continue
            if tag == 'atoms':
                natoms += 1
                atm.append(do_shift[tag](line, shift, resshift))
            if tag == 'bonds':
                bnd.append(do_shift[tag](line, shift, resshift))
            if tag == 'constraints':
                con.append(do_shift[tag](line, shift, resshift))
            if tag == 'angles':
                ang.append(do_shift[tag](line, shift, resshift))
            if tag == 'dihedrals':
                dih.append(do_shift[tag](line, shift, resshift))
            if tag == 'exclusions':
                exc.append(do_shift[tag](line, shift, resshift))
            if tag == 'virtual_sitesn':
                vs.append(do_shift[tag](line, shift, resshift))
            out.append(do_shift[tag](line, shift,resshift))
    return out, natoms, atm, bnd, con, ang, dih, exc, vs
```

```python
nshift = 0
resshift=0
with open('Complex_I.itp', 'w') as out:
    out.write('[ moleculetype ]\nComplex_I 1\n\n[ atoms ]\n')
    for i in range(len(glob.glob('itps/*.itp'))):
        #print(nshift)
        itp, nat, atm, bnd, con, ang, dih, exc, vs = renumber_itp(f'itps/molecule_{i}.itp', nshift, resshift)
        resshift = int(atm[-1].split()[2])
        #print(resshift)
        out.writelines(atm)
        nshift += nat
    nshift = 0
    out.write('\n\n[ bonds ]\n')
    for i in range(len(glob.glob('itps/*.itp'))):
        #print(nshift)
        itp, nat, atm, bnd, con, ang, dih, exc, vs = renumber_itp(f'itps/molecule_{i}.itp', nshift)
        out.writelines(bnd)
        nshift += nat
    nshift = 0
    out.write('\n\n[ constraints ]\n')
    for i in range(len(glob.glob('itps/*.itp'))):
        #print(nshift)
        itp, nat, atm, bnd, con, ang, dih, exc, vs = renumber_itp(f'itps/molecule_{i}.itp', nshift)
        out.writelines(con)
        nshift += nat
    nshift = 0
    out.write('\n\n[ angles ]\n')
    for i in range(len(glob.glob('itps/*.itp'))):
        #print(nshift)
        itp, nat, atm, bnd, con, ang, dih, exc, vs = renumber_itp(f'itps/molecule_{i}.itp', nshift)
        out.writelines(ang)
        nshift += nat
    nshift = 0
    out.write('\n\n[ dihedrals ]\n')
    for i in range(len(glob.glob('itps/*.itp'))):
        #print(nshift)
        itp, nat, atm, bnd, con, ang, dih, exc, vs = renumber_itp(f'itps/molecule_{i}.itp', nshift)
        out.writelines(dih)
        nshift += nat
    nshift = 0
    out.write('\n\n[ exclusions ]\n')
    for i in range(len(glob.glob('itps/*.itp'))):
        #print(nshift)
        itp, nat, atm, bnd, con, ang, dih, exc, vs = renumber_itp(f'itps/molecule_{i}.itp', nshift)
        out.writelines(exc)
        nshift += nat
    nshift = 0
    out.write('\n\n[ virtual_sitesn ]\n')
    for i in range(len(glob.glob('itps/*.itp'))):
        #print(nshift)
        itp, nat, atm, bnd, con, ang, dih, exc, vs = renumber_itp(f'itps/molecule_{i}.itp', nshift)
        out.writelines(vs)
        nshift += nat
```
